```python
#Import required libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
from numpy import asarray
from skimage.io import imread, imshow
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
#Read the image using cv2
img_before=cv2.imread('/content/noise.jpg')
```

```python
#Convert the image from 3 channels to 1 channel by taking the average
img_arr = np.average(img_before,axis=-1)
print(img_arr.shape)
```

### **Mean Filter**

```python
#Input the filter size
filter_size = int(input("Enter Filter Size: "))
```

```python
t = int((filter_size-1)/2)
print(t)
```

```python
#Initializing the output array
output = np.zeros((img_arr.shape[0],img_arr.shape[1]))
print(output.shape)
```

```python
#Applying mean filter to the image
for i in range(t,img_arr.shape[0]-t):
    for j in range(t,img_arr.shape[1]-t):
        temp = img_arr[i-t:i+t+1,j-t:j+t+1]
        output[i][j] = int(np.mean(temp))
      
img_mean = Image.fromarray(output)
```

```python
#Plot the subplots
fig,ax=plt.subplots(1,2,figsize=(15,15))
ax[0].set_title('Original Image')
ax[0].imshow(img_before)

ax[1].set_title('Image after applying mean filter')
ax[1].imshow(img_mean)
```

## **Sharpening**

### **Laplacian Mask**

```python
#Read the image using cv2
img_before_sharpening=cv2.imread('/content/sharpen_img.jpg')
```

```python
img_arr = np.average(img_before,axis=-1)
print(img_arr.shape)
```

**Mask 1**

```python
w1 = [[1,1,1],[1,-8,1],[1,1,1]]
```

```python
sharp_img_shape = np.zeros((img_arr.shape[0],img_arr.shape[1]))
print(sharp_img_shape.shape)
```

```python
for i in range(1,img_before_sharpening.shape[0]-1):
    for j in range(1,img_before_sharpening.shape[1]-1):
        temp = np.dot(img_arr[i-1:i+2,j-1:j+2].flatten(),np.array(w1).flatten())
        sharp_img_shape[i][j] = int(temp)
```

```python
sharp_img_shape+=img_arr 
image_after_sharpening = Image.fromarray(sharp_img_shape)
```

## **Sobel Edge Detection**

```python
s1 = [[-1,-2,-1],[0,0,0],[1,2,1]]
s2 = [[-1,0,1],[-2,0,2],[-1,0,1]]
```

```python
sobel_output1 = np.zeros((img_arr.shape[0],img_arr.shape[1]))
print(sobel_output1.shape)

sobel_output2 = np.zeros((img_arr.shape[0],img_arr.shape[1]))
print(sobel_output2.shape)
```

```python
for i in range(1,img_arr.shape[0]-1):
    for j in range(1,img_arr.shape[1]-1):
        temp = np.dot(img_arr[i-1:i+2,j-1:j+2].flatten(),np.array(s1).flatten())
        sobel_output1[i][j] = int(temp)

for i in range(1,img_arr.shape[0]-1):
    for j in range(1,img_arr.shape[1]-1):
        temp = np.dot(img_arr[i-1:i+2,j-1:j+2].flatten(),np.array(s2).flatten())
        sobel_output2[i][j] = int(temp)
```

```python
sharp_output= img_arr + sobel_output1
output_data_x = Image.fromarray(sharp_output)
```

```python
sharp_output= img_arr + sobel_output2 
output_data_y = Image.fromarray(sharp_output)
```

```python
sharp_output= img_arr + sobel_output1 + sobel_output2 
output_data = Image.fromarray(sharp_output)
```

```python
#Plot the subplots
fig,ax=plt.subplots(1,4,figsize=(20,20))
ax[0].set_title('Original Image')
ax[0].imshow(img_before_sharpening)

ax[1].set_title('Sobel X Output')
ax[1].imshow(output_data_x)

ax[2].set_title('Sobel Y Output')
ax[2].imshow(output_data_y)

ax[3].set_title('Sobel X+Y Output')
ax[3].imshow(output_data)
```
