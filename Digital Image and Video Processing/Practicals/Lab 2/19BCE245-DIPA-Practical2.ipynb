```python
#lev = int(input("Enter the number of intensity levels: "))
def process(lev):
  thresh = 1/(lev-1)
  inter = [0]
  j = thresh
  for i in range(lev-1):
    inter.append(j)
    j += thresh

  row = int(img.shape[0])
  col = int(img.shape[1])
  proc_img = np.zeros((row,col))
  for i in range(row):
    for j in range(col):
      dist = []
      dist = (abs(inter - img[i][j]))
      minim = np.min(dist)
      a = list(dist).index(minim)
      proc_img[i][j] = inter[a]

  return proc_img
    
```

```python
proc_img2 = process(2)
proc_img4 = process(4)
proc_img8 = process(8)
proc_img16 = process(16)
proc_img32 = process(32)
proc_img64 = process(64)
proc_img128 = process(128)
```
