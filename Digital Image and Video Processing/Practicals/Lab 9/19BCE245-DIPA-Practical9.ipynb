```python
import cv2
from PIL import Image as im
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("screenshot.jpg")
img_arr = np.average(img,axis=-1)
print(img_arr.shape)
data = im.fromarray(img_arr)
plt.figure(figsize=(10,10))
plt.imshow(data)
plt.show()
```

# Canny Edge

```python
t_lower = 80  # Lower Threshold
t_upper = 160  # Upper threshold
  
edge = cv2.Canny(img, t_lower, t_upper)
edge_arr = np.asarray(edge)
print(edge_arr.shape)
data_1 = im.fromarray(edge_arr)
plt.figure(figsize=(10,10))
plt.imshow(data_1,cmap="gray")
plt.show()
```

```python
image = cv2.imread('color_img.jpg')
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(img_gray, 150, 255, cv2.THRESH_BINARY)
contours, hierarchy = cv2.findContours(image=thresh, mode=cv2.RETR_EXTERNAL, method=cv2.CHAIN_APPROX_NONE)
                                      
# draw contours on the original image
image_copy = image.copy()
cv2.drawContours(image=image_copy, contours=contours, contourIdx=-1, color=(255, 255, 255), thickness=0, lineType=cv2.LINE_AA)
                
# see the results
plt.imshow(image_copy)
plt.title("None Approximation")
cv2.imwrite('contours_none_image1.jpg', image_copy)
```

```python
z=0
boundary_points = []
for i in contours:
    for j in i:
        boundary_points.append(j[0])
points = np.array(boundary_points)
for i in points:
    print(i)
```

```python
img = cv2.imread('contours_none_image1.jpg')
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
# convert the grayscale image to binary image
ret,thresh = cv2.threshold(gray_image,127,255,0)
M = cv2.moments(thresh)
# calculate x,y coordinate of center
cX = int(M["m10"] / M["m00"])
cY = int(M["m01"] / M["m00"])
print(cX)
print(cY)
```

```python
import math
y = []
x = []
p=0
for i in points:
    dist = math.sqrt(pow(cX-i[0],2) + pow(cY-i[1],2)) 
    y.append(dist)
    x.append(p)
    p+=1    
```

```python
plt.plot(x,y)
plt.show()
```
